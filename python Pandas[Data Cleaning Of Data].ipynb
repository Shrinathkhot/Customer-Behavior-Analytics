```python
# we need to convert all columns name into lower case
df.columns = df.columns.str.lower()
df.columns
```

```python
# now we need to replace the spaces' ' with underscore '_' for better understanding
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
# as we can see a column name 'purchase_amount_(usd)', we need to rename it as 'purchase_amount'
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# making a new column 'age_group' using 'age' column
labels=['Young Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head(10)
```

```python
# creating a column purchase_frequency_days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)

df[['frequency_of_purchases','purchase_frequency_days']].head()
```
